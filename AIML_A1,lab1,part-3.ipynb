```python
import matplotlib.pyplot as plt


people = ['datha','hrushi','charith','sujal']
age = [17,19,18,20]
weights = [50,62,59,58]
height = [150,160,155,165]


plt.scatter(weights, height)
plt.title("Relationship between Weight and Height of Patients")
plt.ylabel("Height in Meters")
plt.xlabel("Weight of Patients in Kgs")
plt.show()

```

```python

import numpy as np
plt.figure(figsize=(7,5))

plt.hist(weights,color='red',edgecolor='k', alpha=0.75,bins=5)

plt.title("Histogram of patient weight",fontsize=18)
plt.xlabel("Weight in kgs",fontsize=15)

plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.title("People's weight in kgs",fontsize=16,
          fontstyle='italic')

plt.bar(x=people,height=weights, width=0.6,
        color='orange',edgecolor='k',alpha=0.6)

plt.xlabel("People",fontsize=15)

plt.ylabel("Weight (in kgs)",fontsize=15)
plt.show()
```
